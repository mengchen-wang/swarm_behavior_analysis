```python
import matplotlib.pyplot as plt
import numpy as np
import math
pi = math.pi
```

```python
def compute_trace(timeRange, stepSize, bx, by, bz, fx, fy, fz, phase_x, phase_y, phase_z, bxc=0, byc=0, bzc=0):
    xlist = []
    ylist = []
    zlist = []
    x = y = z = 0
    for i in range (int(timeRange/stepSize)):
        t = stepSize*i
        x=bxc+bx*np.sin(2*(pi/fx)*t+phase_x)
        xlist.append(x)
        y=byc+by*np.sin(2*(pi/fy)*t+phase_y)
        ylist.append(y)
        z=bzc+bz*np.sin(2*(pi/fz)*t+phase_z)
        zlist.append(z)
    return xlist, ylist, zlist
```

```python
def draw(bx, by, bz, fx, fy, fz, bxc, byc, bzc, phase_x, phase_y, phase_z, sample=False):
    size = 4.5
    ax = plt.axes(projection = '3d')
    ax.view_init(25,35)
    custom_color = (97/255, 124/255, 184/255)
    x, y, z = compute_trace(6, 0.001, bx, by, bz, fx, fy, fz, phase_x, phase_y, phase_z)
    ax.scatter3D(x,y,z, c=[custom_color], s = 1)
   
    ax.set_xlim(-size,size)
    ax.set_ylim(-size,size)
    ax.set_zlim(-size,size)
    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
draw(1, 4, 5, 3, 2, 3, 1, 2, 3.5, 30, 20, 50)
```

```python
def rgb_to_hex(r,g,b):
    # Ensure RGB values are integers in the range 0-255
    # Convert each RGB value to hexadecimal and concatenate them
    hex_color = "#{:02x}{:02x}{:02x}".format(r, g, b)
    return hex_color
rgb_to_hex(63, 59, 114), rgb_to_hex(97, 124, 184), rgb_to_hex(206, 217, 235), rgb_to_hex(223, 155, 146),rgb_to_hex(192, 107, 94)
```
